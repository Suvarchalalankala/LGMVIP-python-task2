```python
from turtle import * 
import time
import random

main =Screen() 
main.setup(width=600,height=600)


main.title("Snake Game by Suvarchala")
main.bgcolor("turquoise")

main.addshape('fruit.gif')
main.addshape('tail.gif')
main.addshape('face.gif')
main.tracer(False)


def move_snake():
    if face.direction=='up':
        y=face.ycor()
        y=y+20
        face.sety(y)
    
    if face.direction=='down':
        y=face.ycor()
        y=y-20
        face.sety(y)
    if face.direction=='right':
        x=face.xcor()
        x=x+20
        face.setx(x)
    if face.direction=='left':
        x=face.xcor()
        x=x-20
        face.setx(x)

def up():
    if face.direction !='down':
        face.direction='up'

def down():
    if face.direction !='up':

        face.direction='down'

def left():
    if face.direction !='right':
        face.direction='left'

def right():
    if face.direction !='left':
        face.direction='right'


# Defining the head of the snake
face=Turtle()
face.shape('face.gif')
face.penup()
face.direction='right'
face.goto(0,160)

#Defining the food of the snake
fruit=Turtle()
fruit.penup()
fruit.shape('fruit.gif')


#Defining the Scores of the game
score=Turtle()
score.penup()
score.goto(0,268)
score.hideturtle()
score.color('black')
score.write('SCORE: ',font=('algerian','25'),align='center')
newscore=0

main.listen()

main.onkeypress(up,'Up')
main.onkeypress(down,'Down')
main.onkeypress(left,'Left')
main.onkeypress(right,'Right')

lose=Turtle()
lose.penup()
lose.hideturtle()

list1=[]
while True:
    main.update()
    if face.xcor()>280 or face.xcor()<-280 or face.ycor()>280 or face.ycor()<-280:
        lose.write('YOU LOST !!',font=('algerian','35'),align='center')
        time.sleep(1)
        lose.clear()
        time.sleep(1)
        face.goto(0,0)
        face.direction='stop'
        for tail in list1:
            tail.goto(10000,10000)
        
        newscore=0
        list1.clear()
        score.clear()
        score.write('SCORE: 0',font=('algerian','25'),align='center')

    
    move_snake()

    for tail in list1:
        if tail.distance(face)<20:
            time.sleep(1)
            face.goto(0,0)
            face.direction='stop'

            for tail in list1:
                tail.goto(1000,1000)
            
            list1.clear()
            newscore=0
            lose.write('Game Lost !!',font=('algerian','25'),align='center')
            time.sleep(1)
            lose.clear()
            score.clear()
            score.write('SCORE: ',font=('algerian','25'),align='center')



    
            for tail in list1:
                tail.goto(1000,1000)

            time.sleep(1)

            list1.clear()
            newscore=0
    

    if face.distance(fruit)<20:
        x=random.randint(-275,275)
        y=random.randint(-275,275)

        fruit.goto(x,y)
        tail=Turtle()
        tail.shape('tail.gif')
        tail.penup()
        list1.append(tail)
        
        newscore=newscore+10
        score.clear()
        score.write(f'SCORE:{newscore}',font=('algerian','25'),align='center')

    time.sleep(.1)


    for i in range(len(list1)-1,0,-1):
            x=list1[i-1].xcor()
            y=list1[i-1].ycor()
            list1[i].goto(x,y)

    if len(list1)>0:
        x=face.xcor()
        y=face.ycor()
        list1[0].goto(x,y)


turtle.done()
```
